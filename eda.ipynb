```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("materials_data_full.csv")

df.head()
```

```python
df[["band_gap","formation_energy"]].describe()
```

```python
import numpy as np

conditions = [
    df["band_gap"] == 0,
    df["band_gap"] <= 4,
    df["band_gap"] > 4
]

choices = ["metal", "semiconductor", "insulator"]

df["band_type"] = np.select(conditions, choices, default='unknown')

counts = df["band_type"].value_counts()
print(counts)
```

```python
fig, ax = plt.subplots()

ax.bar(counts.index, counts.values, color=['red', 'blue', 'green'])
ax.set_xlabel("Material class").set_color("Red")
ax.set_ylabel("Number of materials")
ax.set_title("Dataset composition")

plt.show()
```
